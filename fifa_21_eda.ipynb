```python
import pandas as pd

# Corrected file path
file_path = 'C:/Users/grink/Desktop/FIFA 24 players dataset/player_stats.csv'

# Load the CSV file
fifa = pd.read_csv(file_path, encoding="latin1")
print(fifa.head())
```

```python
##Nation wise Player Representation
nat_cnt=fifa.groupby('country').apply(lambda x:x['player'].count()).reset_index(name='Counts')
nat_cnt.sort_values(by='Counts',ascending=False,inplace=True)
top_20_nat_cnt=nat_cnt[:20]
fig=px.bar(top_20_nat_cnt,x='country',y='Counts',color='Counts',title='Nationwise Representation in the FIFA Game')
fig.show()

#Players from the top 20 most populous nations in the game
```

```python
#Which countries have the most valued players
cnt_best_avg=fifa.groupby('country').apply(lambda x:np.average(x['value'])).reset_index(name='Overall Value')
cnt_best_cnt=fifa.groupby('country').apply(lambda x:x['value'].count()).reset_index(name='Player ')
snt_best_avg_cnt=pd.merge(cnt_best_avg,cnt_best_cnt,how='inner',left_on='country',right_on='country')
sel_best_avg_cnt=snt_best_avg_cnt[snt_best_avg_cnt['Player Values']>=200]
sel_best_avg_cnt.sort_values(by=['Overall Value','Player Values'],ascending=[False,False])
px.scatter(sel_best_avg_cnt,x='Overall Value',y='Player Values',color='Player Values',size='Overall Value',hover_data=['country'],title='Nationwise Player counts and Average Value')
```

```python
# Ensure the 'value' column is properly cleaned
fifa['value'] = (
    fifa['value'].astype(str)  # Ensure the column is treated as string
    .str.replace('$', '', regex=False)  # Remove dollar sign
    .str.replace('.', '', regex=False)  # Remove periods (thousands separator)
    .str.replace(',', '.', regex=False)  # Replace commas with decimal points
    .astype(float)  # Convert to float
)

# Verify the cleaned column
print(fifa['value'].head())

# Proceed with your analysis
cnt_best_avg = fifa.groupby('country').apply(lambda x: np.average(x['value'])).reset_index(name='Overall Value')
cnt_best_cnt = fifa.groupby('country').apply(lambda x: x['value'].count()).reset_index(name='Player Values')
snt_best_avg_cnt = pd.merge(cnt_best_avg, cnt_best_cnt, how='inner', on='country')
sel_best_avg_cnt = snt_best_avg_cnt[snt_best_avg_cnt['Player Values'] >= 200]
sel_best_avg_cnt = sel_best_avg_cnt.sort_values(by=['Overall Value', 'Player Values'], ascending=[False, False])

# Scatter plot
fig = px.scatter(
    sel_best_avg_cnt,
    x='Overall Value',
    y='Player Values',
    color='Player Values',
    size='Overall Value',
    hover_data=['country'],
    title='Nationwise Player Counts and Average Value'
)
fig.show()

#Observation England has the highest number of players, while Brazil has lower players but each have a higher value
```

```python
clb_cnt=fifa.groupby('club').apply(lambda x:x['player'].count()).reset_index(name='Counts')
clb_cnt.sort_values(by='Counts',ascending=False,inplace=True)
fig=px.scatter(clb_cnt,x='club',y='Counts',color='Counts',title='Clubwise Player counts in FIFA 24')
fig.show()

#Clubwise count of Players
```

```python
#Height vs Weight variation
props=fifa[['player','country','club','height','weight']]
fig=px.scatter(props,x='weight',y='height',color='height',size='weight',hover_data=['player','country','club'],title='Height vs Weight Variation of the players in FIFA 24')
fig.show()
```

```python
#Age distribution of players
age_cnt=fifa.groupby('age').apply(lambda x:x['player'].count()).reset_index(name='Counts')
fig=px.bar(age_cnt,x='age',y='Counts',color='Counts',title='Agewise Player distribution in FIFA 24')
fig.show()

#Shows a left skewed function, thus implying most footballers are young in the 20-25 age bracket
```

```python
#Best Players in the dataset based on values

top_play=fifa[['player','value',"age",'club']]
top_play.sort_values(by='value',ascending=False,inplace=True)
top_30_play=top_play[:100]
fig=px.scatter(top_30_play,x='age',y='value',color='age',size='value',hover_data=['player','club'],title='Top Football Players in the FIFA 24 game')
fig.show()

#Mbappe and Haaland are the two most valuable players right now
```

```python
import plotly.express as px
import plotly.graph_objects as go  # Import the 'go' module

# Calculate a weighted score for players
fifa['score'] = (
    fifa['value'] * 0.4 +  # Assign 40% weight to 'value'
    fifa['shot_power'] * 0.2 +  # Assign 20% weight to 'shot_power'
    fifa['long_shots'] * 0.2 +  # Assign 20% weight to 'long_shots'
    fifa['finishing'] * 0.2  # Assign 20% weight to 'finishing'
)

# Sort players by the calculated score in descending order
best_team = fifa.sort_values(by='score', ascending=False)

# Select the top players for the team (e.g., Top 11 for a soccer team)
final_team = best_team[['player', 'club', 'age', 'value', 'shot_power', 'long_shots', 'finishing', 'score']].head(11)

# Create a dynamic table using Plotly
table = go.Figure(data=[go.Table(
    header=dict(values=['<b>Player</b>', '<b>Club</b>', '<b>Age</b>', '<b>Value</b>', '<b>Shot Power</b>', 
                        '<b>Long Shots</b>', '<b>Finishing</b>', '<b>Score</b>'],
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[final_team['player'], final_team['club'], final_team['age'], final_team['value'],
                       final_team['shot_power'], final_team['long_shots'], final_team['finishing'], final_team['score']],
               fill_color='lavender',
               align='left'))
])

# Show the dynamic table
table.show()

# Create a scatter plot for visualizing the top players
scatter_plot = px.scatter(
    final_team,
    x='score',
    y='value',
    size='shot_power',
    color='club',
    hover_data=['player', 'age', 'long_shots', 'finishing'],
    title='Top Players by Score and Value'
)

# Show the scatter plot
scatter_plot.show()


#Shows best Players by their value, shot power, finishing, and long shots capabilities. As you can see: Mbappe, Haaland, KDB, and Mane make the cut in this.
```

```python
#Now lets check the players of Brazil

india = fifa[fifa['country'] == 'India'] 
india.head(10)

#Agewise best players of Brazil

india_team=india

age_play=india_team.groupby('age').apply(lambda x:np.max(x['value'])).reset_index(name='Overall Value')
player_age=pd.merge(india_team,age_play,how='inner',left_on=['age','value'],right_on=['age','Overall Value'])
age_best=player_age[['player','age','Overall Value']]

# Sort by age in ascending order
age_best = age_best.sort_values(by='age', ascending=True)

cm = sns.light_palette("green", as_cmap=True) 
age_best.style.background_gradient(cmap=cm).format(precision= 2)


#Observation shows India's most valued players are in the age group of 22-31. With Anirudh Thapa and Chhangte as the most valued players in the team.
```
